```python
import pandas as pd
import numpy as np
import sqlalchemy
import pandasql as ps
```

```python
df=pd.read_csv('titanic.csv')
```

```python
ps.sqldf("""

SELECT *
FROM df;

""")
```

```python
ps.sqldf("""
SELECT 
df.Embarked,
AVG(df.Age),
SUM(df.Age),
Count(df.PassengerId),
Count(DISTINCT df.Ticket)
FROM df
GROUP BY df.Embarked;
""")
```

```python
ps.sqldf("""
SELECT 
df.Embarked,
AVG(df.Age),
SUM(df.Age),
MAX(df.Age),
MIN(df.Age),
Count(df.PassengerId),
Count(DISTINCT df.Ticket)
FROM df
GROUP BY df.Embarked;
""")
```

```python
ps.sqldf("""
SELECT 
df.Embarked,
df.Sex,
AVG(df.Age),
SUM(df.Age),
MAX(df.Age),
MIN(df.Age),
Count(df.PassengerId),
Count(DISTINCT df.Ticket)
FROM df
GROUP BY df.Embarked,
df.Sex;
""")
```

```python
ps.sqldf("""
SELECT 
df.Embarked,
df.Sex,
AVG(df.Age),
SUM(df.Age),
MAX(df.Age),
MIN(df.Age),
Count(df.PassengerId),
Count(DISTINCT df.Ticket)
FROM df

WHERE Age>=18 
AND Age<=60

GROUP BY df.Embarked,
df.Sex;
""")
```

```python
ps.sqldf("""
SELECT 
df.Embarked,
df.Sex,
AVG(df.Age),
SUM(df.Age),
MAX(df.Age),
MIN(df.Age),
Count(df.PassengerId),
Count(DISTINCT df.Ticket)
FROM df

WHERE Age>=18 
AND Age<=60

GROUP BY df.Embarked,
df.Sex

ORDER BY df.Sex;
""")
```

```python
 ps.sqldf("""
SELECT 
df.Embarked,
df.Sex,
AVG(df.Age),
SUM(df.Age),
MAX(df.Age),
MIN(df.Age),
Count(df.PassengerId),
Count(DISTINCT df.Ticket)
FROM df

WHERE Age>=18 
AND Age<=60

GROUP BY df.Embarked,
df.Sex

ORDER BY df.Sex AND df.Embarked;
""")
```
