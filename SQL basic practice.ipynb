```python
import pandas as pd
import pandasql as ps
```

```python
df=pd.read_csv('titanic.csv')
```

```python
df.head(5)
```

```python
ps.sqldf("""
SELECT
*
FROM df


""")
```

```python
ps.sqldf("""
SELECT
    *
    FROM df AS titanic
    LIMIT 5

""")
```

```python
ps.sqldf("""
SELECT
 PassengerId AS ID
 ,Name AS Name_of_passenger
 FROM df
 LIMIT 2;

""")
```

```python
df1=ps.sqldf("""
  SELECT 
  *
  from df
  WHERE Name='Braund, Mr. Owen Harris';
""")
```

```python
df1
```

```python
ps.sqldf("""
SELECT
* FROM df
WHERE PassengerId BETWEEN 45 AND 50;
""")
```

```python
DSTART=ps.sqldf("""
SELECT * FROM df
WHERE df.Age IS NULL
""")
DSTART.head(5)
```

```python
DSTART=ps.sqldf("""
SELECT * FROM df
WHERE df.Age IS NOT NULL
""")
DSTART.head(5)
```

```python
DSTART=ps.sqldf("""
SELECT * FROM df
WHERE df.Age>18
AND df.Sex='female'
ORDER BY Age
""")
DSTART.head(5)
```

```python
DSTART=ps.sqldf("""
SELECT * FROM df
WHERE df.Cabin IN('A5','A6','A7')
""")
DSTART.head(5)
```

```python
DSTART=ps.sqldf("""
SELECT * FROM df
WHERE Name LIKE 'D%'

""")
DSTART.head(3)
```

```python
DSTART=ps.sqldf("""
SELECT * FROM df
WHERE Name LIKE '%bert%'
""")
DSTART.head()
```

```python
DSTART=ps.sqldf("""
SELECT * FROM df
WHERE Name LIKE '__a%'
""")
DSTART.head()
```

```python
DSTART=ps.sqldf("""
SELECT * FROM df
WHERE Name LIKE 'b%s'OR 'c%'
""")
DSTART.head()
```
